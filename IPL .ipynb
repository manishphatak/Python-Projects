```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(15,7)})
```

```python
ipl= pd.read_csv("IPL2020.csv")
ipl_info= pd.read_csv("IPL2020_info.csv")
ipl_deliveries= pd.read_csv("Deliveries IPL 2020.csv")
ipl_matches=pd.read_csv("Matches IPL 2020.csv")
```
